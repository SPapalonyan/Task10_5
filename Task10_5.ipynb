```
import pandas as pd
from scipy.stats import ttest_ind


def website_analyse(filepath):
    df = pd.read_csv(filepath)

    # 1) Среднее количество просмотренных страниц
    mean_pages = df["Просмотренные страницы"].mean()
    print(f"1) Среднее количество просмотренных страниц: {mean_pages}")

    # 2) Стандартное отклонение времени на сайте
    std_time = df["Время на сайте (сек)"].std()
    print(f"2) Стандартное отклонение: {std_time}")

    # 3) Коэффициент вариации
    mean_time = df["Время на сайте (сек)"].mean()
    cv = std_time / mean_time
    if cv > 0.25:
        variability = "Высокая вариабельность"
    elif cv > 0.1:
        variability = "Умеренная вариабельность"
    else:
        variability = "Слабая вариабельность"
    print(f"3) {variability}: {cv}")

    # 4) t-тест: мобильные vs десктоп
    mobile = df[df["Тип устройства"] == "Мобильное"]["Просмотренные страницы"]
    desktop = df[df["Тип устройства"] == "Десктоп"]["Просмотренные страницы"]
    _, p_value = ttest_ind(mobile, desktop)
    if p_value < 0.05:
        significance = "Есть статистически значимая разница в количестве страниц"
    else:
        significance = "Нет статистически значимой разницы"
    print(f"4) {significance}: {p_value}")


CSV_URL = "https://raw.githubusercontent.com/SPapalonyan/Task10_5/main/website_visits_december_2024.csv"

website_analyse(CSV_URL)




```
